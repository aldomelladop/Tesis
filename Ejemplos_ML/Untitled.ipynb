```python
import numpy as np #Calculo numerico, arrays, matrices, etc
import scipy as sc #funcionalidades tales como tratamiento de imagenes, etc
import matplotlib.pyplot as plt #libreria de visualizacion grafica
```

```python
from sklearn.datasets import load_boston
```

```python
#Cargamos la librería
boston = load_boston()
print(boston.DESCR)
```

Formula de minimizar el error cuadrático medio (MCO)\\

$\beta = (X^{T}X)^{-1}X^{T}Y$

```python
X = np.array(boston.data[:,5])
Y = np.array(boston.target)

plt.scatter(X,Y,alpha=0.3)
#plt.show()

#Añadimos columna de 1s para término independiente
X = np.array([np.ones(506),X]).T

# @ multiplicación matricial
# .T traspuesta
# np.linalg.inv inversa de matriz
B = np.linalg.inv(X.T @ X) @ X.T @ Y

plt.plot([4,9],[B[0]+ B[1]*4, B[0]+ B[1]*9],c = 'red')
plt.show()
```
